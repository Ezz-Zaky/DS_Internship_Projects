# Importing libraries


```
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

Importing the dataset

```
dataset = pd.read_csv('Healthcare_dataset.csv', header=None)
headers = pd.DataFrame(dataset.iloc[0, :].values).T
X = pd.DataFrame(dataset.iloc[1 :, :-1].values)
X.columns=headers.iloc[0,:-1].to_numpy()
y = pd.DataFrame(dataset.iloc[1 :, -1].values)
```

# Encoding

```
# Categorical boolean mask
#categorical_feature_mask = X.dtypes == object
#categorical_cols = X.columns[categorical_feature_mask].tolist()
#categorical_cols.pop()
categorical_cols=X.drop(columns='Dexa_Freq_During_Rx').columns
ohe = OneHotEncoder(sparse=False)
#One-hot-encode the categorical columns.
array_hot_encoded = ohe.fit_transform(X[categorical_cols])
#Convert it to df
X_hot_encoded = pd.DataFrame(array_hot_encoded)
X_hot_encoded.columns = ohe.get_feature_names(X[categorical_cols].columns)
#Extract only the columns that didn't need to be encoded
X_other_cols = X.drop(columns=categorical_cols)
le_y = LabelEncoder()
y = pd.DataFrame(le_y.fit_transform(y)).reset_index(drop=True)
y.columns=[headers.iloc[0,-1]]
```

# Obtaining final dataset 

```
data_out = pd.concat([X_hot_encoded, X_other_cols], axis=1)
data_out = pd.concat([data_out, y], axis=1).reset_index(drop=True)
data_out
```

```
data_out.to_csv('Healthcare_processed_dataset.csv')
```
